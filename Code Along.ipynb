What is the most a customer has spent on a single purchase and what is their customerId?

```python
sql = """SELECT CustomerId, Total
FROM Invoice
ORDER BY Total DESC
LIMIT 1;"""
pd.read_sql_query(sql, engine)
```

```python
sql = '''
SELECT i.customerId, i.Total, c.Country
FROM Invoice i
JOIN Customer c
ON c.customerId = i.customerId
ORDER BY Total DESC
LIMIT 1;'''
pd.read_sql_query(sql, engine)
```
